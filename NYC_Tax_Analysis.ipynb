## Cleaning and Research
#### I took to the internet to find the datasets so that I can find the descriptions, and nuances of the datasets. Some key notes of the EITC dataset are that from the years 2008 and ealier anything above one child is referred to as more than one qualifying child, and from the years 2009 and the number of children is specified up to two qualified children, and anything above is referred to as more than two qualifying children. So for that reason I seperated the dataset into two different Dataframes, one capturing the data from the years 2004-2008, and the other capturing the data from the years 2009-2014. I was also curious as to what the codes found under the 'Notes' column of the dataset meant, so I read up on the codes and provided their definitions below.

## Tax Notes Codes
#### d/ In certain instances, Tax Law prohibits the disclosure of individual taxpayer information. 
#### a/ Amount is less than $500. Reported as zero. For Tax Years 2014 and forward, Credit Amounts Claimed displaying as zero (with Number of Claims greater than zero) actually contain data accurate to three decimal places that become available when the hidden decimal data is obtained via an OData Feed.

```python
#Clean Up and Turn NaN Values into 0's
EITC_claims_df = EITC_claims_df.fillna(0)

#View df
EITC_claims_df
```

```python
#Save F_NYC df
EITC_claims_df.to_csv('./Results/NYC_Full_Year_Resident_EITC_bySize_of_Total_Income_df.csv')
```

```python
#Seperate Useful Columns
F_NYC_Full_Year_Resident_EITC_bySize_of_Total_Income_df = EITC_claims_df[['Tax Year', 'Size of Earned Income', 'Notes', 'Number of Claims, Zero Qualifying Children', 
                                                                          'Credit Amount Claimed, Zero Qualifying Children ($000)', 'Average Credit, Zero Qualifying Children',
                                                                          'Number of Claims, One Qualifying Child', 'Credit Amount Claimed, One Qualifying Child ($000)',
                                                                          'Average Credit, One Qualifying Child', 'Number of Claims, More than One Qualifying Child *',
                                                                          'Credit Amount Claimed, More than One Qualifying Child ($000) *', 'Average Credit, More than One Qualifying Child *',
                                                                          'Number of Claims, Total', 'Credit Amount Claimed, Total ($000)', 'Average Credit, Total'
                                                                         ]]

#View df
F_NYC_Full_Year_Resident_EITC_bySize_of_Total_Income_df
```
